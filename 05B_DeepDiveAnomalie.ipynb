```python
# Phase 4B2: Advanced Anomalie-Vorhersage - MTR Anycast Studie
# ===============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Erweiterte Bibliotheken f√ºr Anomalie-Detection und Prediction
from scipy import stats
from scipy.signal import find_peaks
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from collections import defaultdict, Counter
import networkx as nx

# Time Series Libraries
try:
    from statsmodels.tsa.seasonal import seasonal_decompose
    from statsmodels.tsa.holtwinters import ExponentialSmoothing
    from statsmodels.tsa.arima.model import ARIMA
    STATSMODELS_AVAILABLE = True
except ImportError:
    STATSMODELS_AVAILABLE = False
    print("‚ö†Ô∏è Statsmodels nicht verf√ºgbar - limitierte Zeit-Serie-Analysen")

plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (15, 10)

print("=== PHASE 4B2: ADVANCED ANOMALIE-VORHERSAGE ===")
print("Time-Series-Forecasting, Risk-Scoring & Predictive Failure-Detection")
print("="*75)

# ================================================================
# 1. ERWEITERTE ANOMALIE-KLASSIFIKATION UND FEATURE-ENGINEERING
# ================================================================

def extract_advanced_anomaly_features(df, protocol_name):
    """Extrahiert erweiterte Features f√ºr Anomalie-Detection"""
    print(f"\n1. ERWEITERTE ANOMALIE-FEATURE-EXTRAKTION - {protocol_name}")
    print("-" * 60)
    
    # Service-Klassifikation (konsistent mit vorherigen Phasen)
    SERVICE_MAPPING = {
        # IPv4
        '1.1.1.1': {'name': 'Cloudflare DNS', 'type': 'anycast', 'provider': 'Cloudflare'},
        '8.8.8.8': {'name': 'Google DNS', 'type': 'anycast', 'provider': 'Google'}, 
        '9.9.9.9': {'name': 'Quad9 DNS', 'type': 'anycast', 'provider': 'Quad9'},
        '104.16.123.96': {'name': 'Cloudflare CDN', 'type': 'anycast', 'provider': 'Cloudflare'},
        '2.16.241.219': {'name': 'Akamai CDN', 'type': 'traditional-cdn', 'provider': 'Akamai'},
        '193.99.144.85': {'name': 'Heise', 'type': 'unicast', 'provider': 'Heise'},
        '169.229.128.134': {'name': 'Berkeley NTP', 'type': 'unicast', 'provider': 'UC Berkeley'},
        
        # IPv6
        '2606:4700:4700::1111': {'name': 'Cloudflare DNS', 'type': 'anycast', 'provider': 'Cloudflare'},
        '2001:4860:4860::8888': {'name': 'Google DNS', 'type': 'anycast', 'provider': 'Google'},
        '2620:fe::fe:9': {'name': 'Quad9 DNS', 'type': 'anycast', 'provider': 'Quad9'}, 
        '2606:4700::6810:7b60': {'name': 'Cloudflare CDN', 'type': 'anycast', 'provider': 'Cloudflare'},
        '2a02:26f0:3500:1b::1724:a393': {'name': 'Akamai CDN', 'type': 'traditional-cdn', 'provider': 'Akamai'},
        '2a02:2e0:3fe:1001:7777:772e:2:85': {'name': 'Heise', 'type': 'unicast', 'provider': 'Heise'},
        '2607:f140:ffff:8000:0:8006:0:a': {'name': 'Berkeley NTP', 'type': 'unicast', 'provider': 'UC Berkeley'}
    }
    
    df['service_info'] = df['dst'].map(SERVICE_MAPPING)
    df['service_name'] = df['service_info'].apply(lambda x: x['name'] if x else 'Unknown')
    df['service_type'] = df['service_info'].apply(lambda x: x['type'] if x else 'Unknown')
    df['provider'] = df['service_info'].apply(lambda x: x['provider'] if x else 'Unknown')
    
    # Zeitstempel-Verarbeitung
    df['utctime'] = pd.to_datetime(df['utctime'])
    df['hour'] = df['utctime'].dt.hour
    df['day_of_week'] = df['utctime'].dt.dayofweek
    df['day_of_month'] = df['utctime'].dt.day
    df['week_of_year'] = df['utctime'].dt.isocalendar().week
    
    # Erweiterte Performance-Feature-Extraktion
    anomaly_features = []
    
    print(f"Extrahiere erweiterte Features aus {len(df):,} Messungen...")
    
    processed = 0
    for _, row in df.iterrows():
        try:
            if row['hubs'] is not None and len(row['hubs']) > 0:
                hops = row['hubs']
                
                # Basis-Metriken
                total_hops = len(hops)
                responding_hops = sum(1 for hop in hops if hop and hop.get('Avg', 0) > 0)
                lost_hops = total_hops - responding_hops
                loss_rate = lost_hops / total_hops if total_hops > 0 else 0
                
                # Latenz-Charakteristika
                latencies = [hop.get('Avg', 0) for hop in hops if hop and hop.get('Avg', 0) > 0]
                
                if latencies:
                    final_latency = latencies[-1]
                    min_latency = min(latencies)
                    max_latency = max(latencies)
                    avg_latency = np.mean(latencies)
                    latency_variance = np.var(latencies) if len(latencies) > 1 else 0
                    latency_growth_rate = (max_latency - min_latency) / len(latencies) if len(latencies) > 1 else 0
                    
                    # Latenz-Spr√ºnge (gro√üe √Ñnderungen zwischen aufeinanderfolgenden Hops)
                    latency_jumps = []
                    for i in range(1, len(latencies)):
                        jump = latencies[i] - latencies[i-1]
                        if jump > 0:  # Nur positive Spr√ºnge (Latenz-Erh√∂hung)
                            latency_jumps.append(jump)
                    
                    max_latency_jump = max(latency_jumps) if latency_jumps else 0
                    avg_latency_jump = np.mean(latency_jumps) if latency_jumps else 0
                    
                else:
                    final_latency = max_latency = avg_latency = 0
                    min_latency = latency_variance = latency_growth_rate = 0
                    max_latency_jump = avg_latency_jump = 0
                
                # Jitter-Charakteristika
                jitters = [hop.get('Javg', 0) for hop in hops if hop and hop.get('Javg') is not None and hop.get('Javg', 0) > 0]
                avg_jitter = np.mean(jitters) if jitters else 0
                max_jitter = max(jitters) if jitters else 0
                
                # Packet-Loss-Muster
                packet_losses = [hop.get('Loss%', 0) for hop in hops if hop and hop.get('Loss%') is not None]
                total_packet_loss = sum(packet_losses) / len(packet_losses) if packet_losses else 0
                high_loss_hops = sum(1 for loss in packet_losses if loss > 50)
                
                # ASN-Diversit√§t und Routing-Komplexit√§t
                asns = [hop.get('ASN') for hop in hops if hop and hop.get('ASN') and hop.get('ASN') != 'AS???']
                unique_asns = len(set(asns))
                asn_changes = len(asns) - unique_asns  # Anzahl der ASN-Wiederholungen
                
                # Zeitliche Features
                is_weekend = row['day_of_week'] in [5, 6]  # Samstag, Sonntag
                is_business_hours = 9 <= row['hour'] <= 17
                is_peak_hours = row['hour'] in [8, 9, 17, 18, 19, 20]
                
                # Geografische Features (basierend auf aws_coords aus Phase 4B1)
                aws_coords = {
                    'us-west-1': {'continent': 'North America', 'risk_zone': 'low'},
                    'ca-central-1': {'continent': 'North America', 'risk_zone': 'low'},
                    'eu-central-1': {'continent': 'Europe', 'risk_zone': 'low'},
                    'eu-north-1': {'continent': 'Europe', 'risk_zone': 'medium'},
                    'ap-northeast-1': {'continent': 'Asia', 'risk_zone': 'medium'},
                    'ap-south-1': {'continent': 'Asia', 'risk_zone': 'medium'},
                    'ap-southeast-2': {'continent': 'Oceania', 'risk_zone': 'medium'},
                    'ap-east-1': {'continent': 'Asia', 'risk_zone': 'high'},  # Hong Kong
                    'af-south-1': {'continent': 'Africa', 'risk_zone': 'high'},  # Bekanntes Problem
                    'sa-east-1': {'continent': 'South America', 'risk_zone': 'medium'}
                }
                
                region_info = aws_coords.get(row['region'], {'continent': 'Unknown', 'risk_zone': 'medium'})
                continent = region_info['continent']
                geographic_risk = {'low': 1, 'medium': 2, 'high': 3}[region_info['risk_zone']]
                
                # Anomalie-Score berechnen (Composite-Score)
                anomaly_score = 0
                
                # Latenz-Anomalie (basierend auf Service-Typ-Erwartung)
                expected_latency = {'anycast': 5, 'traditional-cdn': 150, 'unicast': 150}
                service_expected = expected_latency.get(row['service_type'], 100)
                
                if final_latency > service_expected * 3:
                    anomaly_score += 3  # Schwere Latenz-Anomalie
                elif final_latency > service_expected * 1.5:
                    anomaly_score += 1  # Moderate Latenz-Anomalie
                
                # Packet-Loss-Anomalie
                if total_packet_loss > 50:
                    anomaly_score += 3
                elif total_packet_loss > 10:
                    anomaly_score += 1
                
                # Jitter-Anomalie
                if max_jitter > 100:
                    anomaly_score += 2
                elif max_jitter > 50:
                    anomaly_score += 1
                
                # Routing-Anomalie (zu viele Hops f√ºr Service-Typ)
                expected_hops = {'anycast': 10, 'traditional-cdn': 20, 'unicast': 20}
                service_expected_hops = expected_hops.get(row['service_type'], 15)
                
                if total_hops > service_expected_hops * 1.5:
                    anomaly_score += 2
                
                # Geografisches Risiko
                anomaly_score += geographic_risk - 1  # 0-2 zus√§tzliche Punkte
                
                # Klassifikation
                is_anomaly = anomaly_score >= 3
                anomaly_severity = 'normal' if anomaly_score < 2 else ('moderate' if anomaly_score < 4 else 'severe')
                
                anomaly_features.append({
                    'measurement_id': row.get('id', f"{row['region']}_{row['dst']}_{row['utctime']}"),
                    'timestamp': row['utctime'],
                    'region': row['region'],
                    'continent': continent,
                    'service_name': row['service_name'],
                    'service_type': row['service_type'],
                    'provider': row['provider'],
                    
                    # Zeitliche Features
                    'hour': row['hour'],
                    'day_of_week': row['day_of_week'],
                    'day_of_month': row['day_of_month'],
                    'week_of_year': row['week_of_year'],
                    'is_weekend': is_weekend,
                    'is_business_hours': is_business_hours,
                    'is_peak_hours': is_peak_hours,
                    
                    # Performance-Features
                    'final_latency': final_latency,
                    'min_latency': min_latency,
                    'max_latency': max_latency,
                    'avg_latency': avg_latency,
                    'latency_variance': latency_variance,
                    'latency_growth_rate': latency_growth_rate,
                    'max_latency_jump': max_latency_jump,
                    'avg_latency_jump': avg_latency_jump,
                    
                    'avg_jitter': avg_jitter,
                    'max_jitter': max_jitter,
                    'total_packet_loss': total_packet_loss,
                    'high_loss_hops': high_loss_hops,
                    
                    # Routing-Features
                    'total_hops': total_hops,
                    'responding_hops': responding_hops,
                    'lost_hops': lost_hops,
                    'loss_rate': loss_rate,
                    'unique_asns': unique_asns,
                    'asn_changes': asn_changes,
                    
                    # Geo-Risk-Features
                    'geographic_risk': geographic_risk,
                    
                    # Anomalie-Klassifikation
                    'anomaly_score': anomaly_score,
                    'is_anomaly': is_anomaly,
                    'anomaly_severity': anomaly_severity
                })
        
        except Exception as e:
            continue
        
        processed += 1
        if processed % 25000 == 0:
            print(f"  Features extrahiert: {processed:,} Messungen...")
    
    features_df = pd.DataFrame(anomaly_features)
    
    if len(features_df) > 0:
        print(f"‚úì {len(features_df):,} erweiterte Feature-Vektoren erstellt")
        
        # Feature-Statistiken
        print(f"\nüìä ANOMALIE-VERTEILUNG:")
        anomaly_count = features_df['is_anomaly'].sum()
        print(f"  Gesamte Anomalien: {anomaly_count:,} ({anomaly_count/len(features_df)*100:.1f}%)")
        
        severity_counts = features_df['anomaly_severity'].value_counts()
        for severity, count in severity_counts.items():
            print(f"  {severity}: {count:,} ({count/len(features_df)*100:.1f}%)")
        
        return features_df
    
    else:
        print("‚ùå Keine Features extrahiert")
        return None

# ================================================================
# 2. TIME-SERIES-ANOMALIE-DETECTION
# ================================================================

def perform_time_series_anomaly_detection(features_df, protocol_name):
    """F√ºhrt zeitreihen-basierte Anomalie-Detection durch"""
    print(f"\n2. TIME-SERIES-ANOMALIE-DETECTION - {protocol_name}")
    print("-" * 55)
    
    if features_df is None or len(features_df) == 0:
        print("Keine Features f√ºr Time-Series-Analyse verf√ºgbar")
        return None
    
    # Zeitreihen-Aggregation (st√ºndlich)
    features_df['timestamp_hour'] = features_df['timestamp'].dt.floor('H')
    
    time_series_data = []
    
    print(f"üïí ZEITREIHEN-AGGREGATION UND TREND-ANALYSE:")
    
    # Aggregiere pro Service, Region und Stunde
    for (service, region), group in features_df.groupby(['service_name', 'region']):
        hourly_agg = group.groupby('timestamp_hour').agg({
            'final_latency': ['mean', 'std', 'max'],
            'total_packet_loss': 'mean',
            'avg_jitter': 'mean',
            'total_hops': 'mean',
            'loss_rate': 'mean',
            'anomaly_score': ['mean', 'max'],
            'is_anomaly': ['sum', 'count']
        }).reset_index()
        
        # Flatten column names
        hourly_agg.columns = ['_'.join(col).strip() if col[1] else col[0] for col in hourly_agg.columns]
        
        # Berechne Anomalie-Rate
        hourly_agg['anomaly_rate'] = hourly_agg['is_anomaly_sum'] / hourly_agg['is_anomaly_count']
        
        # F√ºge Metadaten hinzu
        hourly_agg['service_name'] = service
        hourly_agg['region'] = region
        
        time_series_data.append(hourly_agg)
    
    if not time_series_data:
        print("Keine Zeitreihen-Daten erstellt")
        return None
    
    ts_df = pd.concat(time_series_data, ignore_index=True)
    ts_df['timestamp_hour'] = pd.to_datetime(ts_df['timestamp_hour'])
    
    print(f"Zeitreihen-Daten erstellt: {len(ts_df):,} Stunden-Datenpunkte")
    
    # Trend-Detection f√ºr kritische Services
    print(f"\nüìà TREND-DETECTION F√úR KRITISCHE SERVICES:")
    
    critical_services = ['Cloudflare DNS', 'Google DNS', 'Quad9 DNS']
    trend_results = []
    
    for service in critical_services:
        service_data = ts_df[ts_df['service_name'] == service].copy()
        
        if len(service_data) > 24:  # Mindestens 24 Stunden Daten
            service_data = service_data.sort_values('timestamp_hour')
            
            # Trend-Berechnung f√ºr verschiedene Metriken
            metrics = ['final_latency_mean', 'anomaly_rate', 'total_packet_loss_mean']
            
            for metric in metrics:
                if metric in service_data.columns:
                    # Linear trend (Regression)
                    x = np.arange(len(service_data))
                    y = service_data[metric].values
                    
                    # Entferne NaN-Werte
                    valid_mask = ~np.isnan(y)
                    if valid_mask.sum() > 10:  # Mindestens 10 g√ºltige Punkte
                        slope, intercept, r_value, p_value, std_err = stats.linregress(x[valid_mask], y[valid_mask])
                        
                        trend_direction = 'steigend' if slope > 0 else 'fallend'
                        trend_strength = abs(r_value)
                        
                        trend_results.append({
                            'service': service,
                            'metric': metric,
                            'slope': slope,
                            'r_squared': r_value**2,
                            'p_value': p_value,
                            'trend_direction': trend_direction,
                            'trend_strength': trend_strength,
                            'is_significant': p_value < 0.05
                        })
    
    # Trend-Ergebnisse ausgeben
    trend_df = pd.DataFrame(trend_results)
    
    if len(trend_df) > 0:
        significant_trends = trend_df[trend_df['is_significant'] & (trend_df['trend_strength'] > 0.3)]
        
        print(f"Signifikante Trends gefunden: {len(significant_trends)}")
        
        for _, trend in significant_trends.iterrows():
            print(f"  {trend['service']} - {trend['metric']}: {trend['trend_direction']} "
                  f"(R¬≤={trend['r_squared']:.3f}, p={trend['p_value']:.3f})")
            
            if trend['trend_direction'] == 'steigend' and 'latency' in trend['metric']:
                print(f"    ‚ö†Ô∏è Performance-Degradation detektiert!")
            elif trend['trend_direction'] == 'steigend' and 'anomaly' in trend['metric']:
                print(f"    üö® Zunehmende Instabilit√§t!")
    
    # LSTM-basierte Anomalie-Detection (vereinfacht ohne echte LSTM-Bibliothek)
    print(f"\nü§ñ SEQUENCE-BASED ANOMALIE-DETECTION:")
    
    # Verwende Moving-Window-Statistiken als LSTM-Alternative
    window_size = 24  # 24 Stunden
    
    sequence_anomalies = []
    
    for service in critical_services:
        service_ts = ts_df[ts_df['service_name'] == service].copy().sort_values('timestamp_hour')
        
        if len(service_ts) > window_size:
            # Rolling Statistics f√ºr Anomalie-Detection
            service_ts['latency_rolling_mean'] = service_ts['final_latency_mean'].rolling(window=window_size).mean()
            service_ts['latency_rolling_std'] = service_ts['final_latency_mean'].rolling(window=window_size).std()
            
            # Anomalie = Werte au√üerhalb 2.5 Standardabweichungen
            service_ts['upper_bound'] = service_ts['latency_rolling_mean'] + 2.5 * service_ts['latency_rolling_std']
            service_ts['lower_bound'] = service_ts['latency_rolling_mean'] - 2.5 * service_ts['latency_rolling_std']
            
            # Sequence-Anomalien identifizieren
            sequence_anomaly_mask = (
                (service_ts['final_latency_mean'] > service_ts['upper_bound']) |
                (service_ts['final_latency_mean'] < service_ts['lower_bound'])
            )
            
            anomaly_periods = service_ts[sequence_anomaly_mask]
            
            if len(anomaly_periods) > 0:
                for _, anomaly in anomaly_periods.iterrows():
                    sequence_anomalies.append({
                        'service': service,
                        'timestamp': anomaly['timestamp_hour'],
                        'actual_latency': anomaly['final_latency_mean'],
                        'expected_range': f"{anomaly['lower_bound']:.1f}-{anomaly['upper_bound']:.1f}ms",
                        'deviation_factor': abs(anomaly['final_latency_mean'] - anomaly['latency_rolling_mean']) / anomaly['latency_rolling_std']
                    })
    
    if sequence_anomalies:
        seq_anomalies_df = pd.DataFrame(sequence_anomalies)
        print(f"Sequence-Anomalien detektiert: {len(seq_anomalies_df)}")
        
        # Top-Anomalien
        top_anomalies = seq_anomalies_df.nlargest(5, 'deviation_factor')
        print("\nTop-5 Sequence-Anomalien:")
        for _, anomaly in top_anomalies.iterrows():
            print(f"  {anomaly['service']} @ {anomaly['timestamp']}: "
                  f"{anomaly['actual_latency']:.1f}ms ({anomaly['deviation_factor']:.1f}œÉ)")
    
    return ts_df, trend_df, sequence_anomalies

# ================================================================
# 3. MACHINE LEARNING ANOMALIE-PREDICTION-MODELLE
# ================================================================

def build_anomaly_prediction_models(features_df, protocol_name):
    """Baut Machine Learning Modelle f√ºr Anomalie-Vorhersage"""
    print(f"\n3. ML ANOMALIE-PREDICTION-MODELLE - {protocol_name}")
    print("-" * 50)
    
    if features_df is None or len(features_df) == 0:
        print("Keine Features f√ºr ML-Modelling verf√ºgbar")
        return None
    
    # Feature-Selection f√ºr ML
    numeric_features = [
        'final_latency', 'avg_latency', 'latency_variance', 'max_latency_jump',
        'avg_jitter', 'max_jitter', 'total_packet_loss', 'total_hops',
        'loss_rate', 'unique_asns', 'geographic_risk', 'hour', 'day_of_week'
    ]
    
    # Boolean Features zu Numeric
    boolean_features = ['is_weekend', 'is_business_hours', 'is_peak_hours']
    for feat in boolean_features:
        features_df[feat] = features_df[feat].astype(int)
        numeric_features.append(feat)
    
    # Kategorische Features zu Numeric (Label Encoding)
    features_df['service_type_numeric'] = features_df['service_type'].map({
        'anycast': 0, 'traditional-cdn': 1, 'unicast': 2
    })
    features_df['continent_numeric'] = features_df['continent'].map({
        'Europe': 0, 'North America': 1, 'Asia': 2, 'Oceania': 3, 'Africa': 4, 'South America': 5
    })
    
    numeric_features.extend(['service_type_numeric', 'continent_numeric'])
    
    # Entferne Zeilen mit fehlenden Werten
    ml_data = features_df.dropna(subset=numeric_features + ['is_anomaly'])
    
    if len(ml_data) == 0:
        print("Keine sauberen Daten f√ºr ML verf√ºgbar")
        return None
    
    X = ml_data[numeric_features]
    y = ml_data['is_anomaly']
    
    print(f"ML-Dataset: {len(X):,} Samples, {len(numeric_features)} Features")
    print(f"Anomalie-Balance: {y.sum():,} Anomalien ({y.mean():.1%})")
    
    # Feature-Scaling
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Train-Test-Split
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.2, random_state=42, stratify=y
    )
    
    print(f"\nü§ñ ANOMALIE-PREDICTION-MODELLE:")
    
    # 1. Isolation Forest (Unsupervised)
    print(f"\n1. ISOLATION FOREST (Unsupervised Anomalie-Detection):")
    
    iso_forest = IsolationForest(
        contamination=y.mean(),  # Erwartete Anomalie-Rate
        random_state=42,
        n_estimators=100
    )
    
    iso_forest.fit(X_train)
    
    # Predictions (-1 f√ºr Anomalie, 1 f√ºr normal)
    iso_predictions = iso_forest.predict(X_test)
    iso_predictions_binary = (iso_predictions == -1).astype(int)
    
    iso_accuracy = (iso_predictions_binary == y_test).mean()
    iso_precision = (iso_predictions_binary & y_test).sum() / iso_predictions_binary.sum() if iso_predictions_binary.sum() > 0 else 0
    iso_recall = (iso_predictions_binary & y_test).sum() / y_test.sum() if y_test.sum() > 0 else 0
    
    print(f"  Accuracy: {iso_accuracy:.3f}")
    print(f"  Precision: {iso_precision:.3f}")
    print(f"  Recall: {iso_recall:.3f}")
    
    # 2. Random Forest Classifier (Supervised)
    print(f"\n2. RANDOM FOREST CLASSIFIER (Supervised):")
    
    rf_model = RandomForestClassifier(
        n_estimators=100,
        max_depth=10,
        random_state=42,
        class_weight='balanced'  # Handle imbalanced data
    )
    
    rf_model.fit(X_train, y_train)
    
    # Predictions und Evaluation
    rf_predictions = rf_model.predict(X_test)
    rf_proba = rf_model.predict_proba(X_test)[:, 1]
    
    rf_accuracy = (rf_predictions == y_test).mean()
    rf_precision = (rf_predictions & y_test).sum() / rf_predictions.sum() if rf_predictions.sum() > 0 else 0
    rf_recall = (rf_predictions & y_test).sum() / y_test.sum() if y_test.sum() > 0 else 0
    
    try:
        rf_auc = roc_auc_score(y_test, rf_proba)
        print(f"  AUC-ROC: {rf_auc:.3f}")
    except:
        rf_auc = 0
        print(f"  AUC-ROC: Berechnung fehlgeschlagen")
    
    print(f"  Accuracy: {rf_accuracy:.3f}")
    print(f"  Precision: {rf_precision:.3f}")
    print(f"  Recall: {rf_recall:.3f}")
    
    # Feature-Wichtigkeit
    feature_importance = pd.DataFrame({
        'feature': numeric_features,
        'importance': rf_model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    print(f"\n  Top-5 wichtigste Features:")
    for _, row in feature_importance.head(5).iterrows():
        print(f"    {row['feature']}: {row['importance']:.3f}")
    
    # 3. DBSCAN Clustering (Density-based Anomalie-Detection)
    print(f"\n3. DBSCAN CLUSTERING (Density-based):")
    
    dbscan = DBSCAN(eps=0.5, min_samples=10)
    dbscan_labels = dbscan.fit_predict(X_test)
    
    # Anomalien = Punkte im Cluster -1 (Noise)
    dbscan_anomalies = (dbscan_labels == -1)
    
    if dbscan_anomalies.sum() > 0:
        dbscan_precision = (dbscan_anomalies & y_test).sum() / dbscan_anomalies.sum()
        dbscan_recall = (dbscan_anomalies & y_test).sum() / y_test.sum() if y_test.sum() > 0 else 0
        
        print(f"  Anomalien detektiert: {dbscan_anomalies.sum()}")
        print(f"  Precision: {dbscan_precision:.3f}")
        print(f"  Recall: {dbscan_recall:.3f}")
        print(f"  Cluster gefunden: {len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)}")
    else:
        print(f"  Keine Anomalien durch DBSCAN detektiert")
    
    # 4. Ensemble-Prediction (kombiniert alle Modelle)
    print(f"\n4. ENSEMBLE-PREDICTION:")
    
    # Kombiniere Predictions mit Gewichtung
    ensemble_score = (
        0.3 * iso_predictions_binary +
        0.5 * rf_predictions +
        0.2 * dbscan_anomalies.astype(int)
    )
    
    # Threshold f√ºr finale Entscheidung
    ensemble_predictions = (ensemble_score > 0.5).astype(int)
    
    ensemble_accuracy = (ensemble_predictions == y_test).mean()
    ensemble_precision = (ensemble_predictions & y_test).sum() / ensemble_predictions.sum() if ensemble_predictions.sum() > 0 else 0
    ensemble_recall = (ensemble_predictions & y_test).sum() / y_test.sum() if y_test.sum() > 0 else 0
    
    print(f"  Ensemble Accuracy: {ensemble_accuracy:.3f}")
    print(f"  Ensemble Precision: {ensemble_precision:.3f}")
    print(f"  Ensemble Recall: {ensemble_recall:.3f}")
    
    # Modell-Performance-Vergleich
    model_performance = pd.DataFrame({
        'Model': ['Isolation Forest', 'Random Forest', 'DBSCAN', 'Ensemble'],
        'Accuracy': [iso_accuracy, rf_accuracy, np.nan, ensemble_accuracy],
        'Precision': [iso_precision, rf_precision, dbscan_precision if 'dbscan_precision' in locals() else np.nan, ensemble_precision],
        'Recall': [iso_recall, rf_recall, dbscan_recall if 'dbscan_recall' in locals() else np.nan, ensemble_recall]
    })
    
    print(f"\nüìä MODELL-PERFORMANCE-VERGLEICH:")
    print(model_performance.round(3))
    
    return {
        'models': {
            'isolation_forest': iso_forest,
            'random_forest': rf_model,
            'dbscan': dbscan,
            'scaler': scaler
        },
        'feature_names': numeric_features,
        'feature_importance': feature_importance,
        'model_performance': model_performance,
        'test_predictions': {
            'rf_proba': rf_proba,
            'ensemble_score': ensemble_score
        }
    }

# ================================================================
# 4. GEOGRAFISCHES RISK-SCORING-SYSTEM
# ================================================================

def build_geographic_risk_scoring(features_df, protocol_name):
    """Entwickelt geografisches Risk-Scoring-System"""
    print(f"\n4. GEOGRAFISCHES RISK-SCORING-SYSTEM - {protocol_name}")
    print("-" * 55)
    
    if features_df is None or len(features_df) == 0:
        print("Keine Features f√ºr Risk-Scoring verf√ºgbar")
        return None
    
    print(f"üåç REGIONALES RISK-ASSESSMENT:")
    
    # Regional Risk-Scores berechnen
    risk_metrics = []
    
    for region in features_df['region'].unique():
        region_data = features_df[features_df['region'] == region]
        
        if len(region_data) > 100:  # Mindestens 100 Messungen f√ºr statistisch sinnvolle Aussagen
            # Basis-Risk-Metriken
            anomaly_rate = region_data['is_anomaly'].mean()
            avg_anomaly_score = region_data['anomaly_score'].mean()
            severe_anomaly_rate = (region_data['anomaly_severity'] == 'severe').mean()
            
            # Performance-Risk-Indikatoren
            high_latency_rate = (region_data['final_latency'] > 100).mean()
            high_loss_rate = (region_data['total_packet_loss'] > 10).mean()
            routing_complexity = region_data['total_hops'].mean()
            
            # Zeitliche Instabilit√§t
            latency_cv = region_data['final_latency'].std() / region_data['final_latency'].mean() if region_data['final_latency'].mean() > 0 else 0
            
            # Service-spezifische Risks
            anycast_performance = region_data[region_data['service_type'] == 'anycast']['final_latency'].mean()
            anycast_expected = 5  # ms
            anycast_degradation = max(0, (anycast_performance - anycast_expected) / anycast_expected) if anycast_performance else 0
            
            # Composite Risk-Score (0-100)
            risk_score = (
                anomaly_rate * 25 +                    # 25% Gewichtung f√ºr Anomalie-Rate
                severe_anomaly_rate * 20 +             # 20% f√ºr schwere Anomalien
                high_latency_rate * 15 +               # 15% f√ºr hohe Latenzen
                high_loss_rate * 15 +                  # 15% f√ºr Packet Loss
                min(latency_cv * 10, 10) +             # 10% f√ºr Instabilit√§t (gecapped)
                min(anycast_degradation * 10, 10) +    # 10% f√ºr Anycast-Degradation
                min((routing_complexity - 10) / 20 * 5, 5)  # 5% f√ºr Routing-Komplexit√§t
            )
            
            # Risk-Kategorisierung
            if risk_score < 20:
                risk_category = 'Low Risk'
                risk_color = 'üü¢'
            elif risk_score < 40:
                risk_category = 'Medium Risk'
                risk_color = 'üü°'
            elif risk_score < 60:
                risk_category = 'High Risk'
                risk_color = 'üü†'
            else:
                risk_category = 'Critical Risk'
                risk_color = 'üî¥'
            
            # Kontinent-Info
            continent = region_data['continent'].iloc[0] if len(region_data) > 0 else 'Unknown'
            
            risk_metrics.append({
                'region': region,
                'continent': continent,
                'measurements': len(region_data),
                'anomaly_rate': anomaly_rate,
                'severe_anomaly_rate': severe_anomaly_rate,
                'high_latency_rate': high_latency_rate,
                'high_loss_rate': high_loss_rate,
                'anycast_performance': anycast_performance,
                'routing_complexity': routing_complexity,
                'latency_cv': latency_cv,
                'risk_score': risk_score,
                'risk_category': risk_category,
                'risk_color': risk_color
            })
    
    risk_df = pd.DataFrame(risk_metrics).sort_values('risk_score', ascending=False)
    
    print(f"Regionale Risk-Scores:")
    for _, risk in risk_df.iterrows():
        print(f"  {risk['risk_color']} {risk['region']} ({risk['continent']}): "
              f"{risk['risk_score']:.1f}/100 - {risk['risk_category']}")
        print(f"     Anomalie-Rate: {risk['anomaly_rate']:.1%}, "
              f"Anycast: {risk['anycast_performance']:.1f}ms")
    
    # Kontinentales Risk-Ranking
    print(f"\nüåç KONTINENTALES RISK-RANKING:")
    
    continental_risk = risk_df.groupby('continent').agg({
        'risk_score': 'mean',
        'anomaly_rate': 'mean',
        'anycast_performance': 'mean'
    }).sort_values('risk_score', ascending=False)
    
    for continent, stats in continental_risk.iterrows():
        if stats['risk_score'] < 30:
            continent_color = 'üü¢'
        elif stats['risk_score'] < 50:
            continent_color = 'üü°'
        else:
            continent_color = 'üî¥'
        
        print(f"  {continent_color} {continent}: {stats['risk_score']:.1f}/100 Risk-Score")
        print(f"     Anomalie-Rate: {stats['anomaly_rate']:.1%}, "
              f"Anycast-Performance: {stats['anycast_performance']:.1f}ms")
    
    # Provider Risk-Assessment
    print(f"\nüè¢ PROVIDER RISK-ASSESSMENT:")
    
    provider_risk = []
    
    for provider in features_df['provider'].unique():
        provider_data = features_df[features_df['provider'] == provider]
        
        if len(provider_data) > 1000:  # Mindestens 1000 Messungen
            provider_anomaly_rate = provider_data['is_anomaly'].mean()
            provider_avg_latency = provider_data['final_latency'].mean()
            provider_consistency = 1 - (provider_data['final_latency'].std() / provider_avg_latency) if provider_avg_latency > 0 else 0
            
            # Provider-spezifischer Risk-Score
            provider_risk_score = provider_anomaly_rate * 50 + (1 - max(0, provider_consistency)) * 50
            
            provider_risk.append({
                'provider': provider,
                'measurements': len(provider_data),
                'anomaly_rate': provider_anomaly_rate,
                'avg_latency': provider_avg_latency,
                'consistency': provider_consistency,
                'risk_score': provider_risk_score
            })
    
    provider_risk_df = pd.DataFrame(provider_risk).sort_values('risk_score', ascending=False)
    
    print(f"Provider Risk-Ranking:")
    for _, provider in provider_risk_df.iterrows():
        if provider['risk_score'] < 20:
            provider_color = 'üü¢'
        elif provider['risk_score'] < 40:
            provider_color = 'üü°'
        else:
            provider_color = 'üî¥'
        
        print(f"  {provider_color} {provider['provider']}: {provider['risk_score']:.1f}/100")
        print(f"     Anomalie-Rate: {provider['anomaly_rate']:.1%}, "
              f"Latenz: {provider['avg_latency']:.1f}ms")
    
    return {
        'regional_risk': risk_df,
        'continental_risk': continental_risk,
        'provider_risk': provider_risk_df
    }

# ================================================================
# 5. REAL-TIME ANOMALIE-DETECTION-PIPELINE
# ================================================================

def design_realtime_anomaly_pipeline(ml_models, risk_scores, protocol_name):
    """Entwirft Real-Time Anomalie-Detection-Pipeline"""
    print(f"\n5. REAL-TIME ANOMALIE-DETECTION-PIPELINE - {protocol_name}")
    print("-" * 60)
    
    if ml_models is None or risk_scores is None:
        print("Unvollst√§ndige Modelle f√ºr Real-Time-Pipeline")
        return None
    
    print(f"‚ö° REAL-TIME PIPELINE-ARCHITEKTUR:")
    
    pipeline_components = {
        'data_ingestion': {
            'description': 'MTR-Messungen alle 15 Minuten',
            'latency': '<1 Sekunde',
            'throughput': '~100 Messungen/Minute'
        },
        'feature_extraction': {
            'description': 'Echtzeit-Feature-Engineering',
            'latency': '<5 Sekunden',
            'features': len(ml_models['feature_names'])
        },
        'anomaly_scoring': {
            'description': 'ML-Modell-Ensemble-Scoring',
            'models': ['Isolation Forest', 'Random Forest'],
            'latency': '<2 Sekunden'
        },
        'risk_assessment': {
            'description': 'Geografisches Risk-Scoring',
            'risk_levels': ['Low', 'Medium', 'High', 'Critical'],
            'update_frequency': 'St√ºndlich'
        },
        'alerting': {
            'description': 'Multi-Level-Alert-System',
            'channels': ['Dashboard', 'Email', 'Slack', 'SMS'],
            'thresholds': {
                'info': 'Anomalie-Score > 3',
                'warning': 'Anomalie-Score > 5',
                'critical': 'Anomalie-Score > 7'
            }
        }
    }
    
    for component, config in pipeline_components.items():
        print(f"\n  üì¶ {component.upper().replace('_', ' ')}:")
        for key, value in config.items():
            if isinstance(value, list):
                print(f"    {key}: {', '.join(map(str, value))}")
            elif isinstance(value, dict):
                print(f"    {key}:")
                for sub_key, sub_value in value.items():
                    print(f"      {sub_key}: {sub_value}")
            else:
                print(f"    {key}: {value}")
    
    # Simulations-Framework f√ºr Real-Time-Detection
    print(f"\nüîÆ ANOMALIE-PREDICTION-SIMULATION:")
    
    def simulate_realtime_detection(new_measurement):
        """Simuliert Real-Time Anomalie-Detection f√ºr neue Messung"""
        
        # Feature-Extraktion (vereinfacht)
        features = [
            new_measurement.get('latency', 50),
            new_measurement.get('jitter', 5),
            new_measurement.get('packet_loss', 0),
            new_measurement.get('hops', 12),
            new_measurement.get('geographic_risk', 1),
            new_measurement.get('hour', 12),
            new_measurement.get('is_weekend', 0)
        ]
        
        # Erweitere auf alle Features (mit Standardwerten)
        while len(features) < len(ml_models['feature_names']):
            features.append(0)
        
        features_scaled = ml_models['models']['scaler'].transform([features])
        
        # ML-Prediction
        rf_prob = ml_models['models']['random_forest'].predict_proba(features_scaled)[0][1]
        iso_anomaly = ml_models['models']['isolation_forest'].predict(features_scaled)[0] == -1
        
        # Ensemble-Score
        ensemble_score = 0.7 * rf_prob + 0.3 * int(iso_anomaly)
        
        # Risk-Assessment
        region = new_measurement.get('region', 'unknown')
        regional_risk = 0.5  # Default
        
        if risk_scores and 'regional_risk' in risk_scores:
            region_risk_data = risk_scores['regional_risk'][
                risk_scores['regional_risk']['region'] == region
            ]
            if len(region_risk_data) > 0:
                regional_risk = region_risk_data.iloc[0]['risk_score'] / 100
        
        # Finale Anomalie-Klassifikation
        final_score = ensemble_score * 0.7 + regional_risk * 0.3
        
        if final_score > 0.7:
            alert_level = 'CRITICAL'
            action = 'Immediate investigation required'
        elif final_score > 0.5:
            alert_level = 'WARNING'
            action = 'Monitor closely'
        elif final_score > 0.3:
            alert_level = 'INFO'
            action = 'Log for analysis'
        else:
            alert_level = 'NORMAL'
            action = 'No action required'
        
        return {
            'anomaly_probability': ensemble_score,
            'regional_risk': regional_risk,
            'final_score': final_score,
            'alert_level': alert_level,
            'recommended_action': action,
            'model_confidence': abs(0.5 - ensemble_score) * 2  # 0-1 Scale
        }
    
    # Beispiel-Simulationen
    test_scenarios = [
        {'name': 'Normal Cloudflare', 'latency': 2, 'region': 'eu-central-1', 'hops': 8},
        {'name': 'Afrika Problem', 'latency': 50, 'region': 'af-south-1', 'hops': 15, 'geographic_risk': 3},
        {'name': 'Severe Anomalie', 'latency': 500, 'packet_loss': 50, 'hops': 25},
        {'name': 'Akamai Normal', 'latency': 150, 'region': 'us-west-1', 'hops': 18}
    ]
    
    print(f"\nBeispiel-Simulationen:")
    for scenario in test_scenarios:
        result = simulate_realtime_detection(scenario)
        print(f"  üìä {scenario['name']}:")
        print(f"    Final Score: {result['final_score']:.3f}")
        print(f"    Alert Level: {result['alert_level']}")
        print(f"    Action: {result['recommended_action']}")
        print(f"    Confidence: {result['model_confidence']:.1%}")
    
    # Performance-Monitoring-Metriken
    print(f"\nüìà PIPELINE-PERFORMANCE-MONITORING:")
    
    monitoring_metrics = {
        'prediction_accuracy': 'T√§gliche Validation gegen Ground Truth',
        'false_positive_rate': 'Ziel: <5% False Positives',
        'detection_latency': 'Ziel: <10 Sekunden End-to-End',
        'model_drift': 'W√∂chentliche Feature-Drift-Detection',
        'data_quality': 'Missing/Invalid Data Rate <1%',
        'alert_fatigue': 'Critical Alerts <10/Tag'
    }
    
    for metric, target in monitoring_metrics.items():
        print(f"  ‚Ä¢ {metric}: {target}")
    
    return {
        'pipeline_architecture': pipeline_components,
        'simulation_function': simulate_realtime_detection,
        'monitoring_metrics': monitoring_metrics
    }

# ================================================================
# 6. ANOMALIE-PREDICTION VISUALISIERUNGEN
# ================================================================

def create_anomaly_prediction_visualizations(features_df, ml_models, risk_scores, protocol_name):
    """Erstellt umfassende Visualisierungen f√ºr Anomalie-Prediction"""
    print(f"\n6. ANOMALIE-PREDICTION VISUALISIERUNGEN - {protocol_name}")
    print("-" * 60)
    
    if features_df is None:
        print("Keine Daten f√ºr Visualisierungen")
        return
    
    fig, axes = plt.subplots(3, 2, figsize=(20, 18))
    fig.suptitle(f'Advanced Anomalie-Vorhersage - {protocol_name}', fontsize=16, fontweight='bold')
    
    # 1. Anomalie-Score-Verteilung nach Service-Typ
    ax = axes[0, 0]
    
    service_types = features_df['service_type'].unique()
    anomaly_scores_by_type = [
        features_df[features_df['service_type'] == st]['anomaly_score'].values 
        for st in service_types
    ]
    
    bp = ax.boxplot(anomaly_scores_by_type, labels=service_types, patch_artist=True)
    colors = ['lightgreen', 'orange', 'lightcoral']
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)
    
    ax.set_title('Anomalie-Score-Verteilung nach Service-Typ', fontweight='bold')
    ax.set_ylabel('Anomalie-Score')
    ax.tick_params(axis='x', rotation=45)
    ax.grid(True, alpha=0.3)
    
    # 2. Feature-Wichtigkeit (falls ML-Modelle verf√ºgbar)
    ax = axes[0, 1]
    
    if ml_models and 'feature_importance' in ml_models:
        feature_imp = ml_models['feature_importance'].head(10)
        
        bars = ax.barh(range(len(feature_imp)), feature_imp['importance'])
        ax.set_yticks(range(len(feature_imp)))
        ax.set_yticklabels(feature_imp['feature'])
        ax.set_xlabel('Feature-Wichtigkeit')
        ax.set_title('Top-10 Features f√ºr Anomalie-Prediction', fontweight='bold')
        ax.grid(True, alpha=0.3)
        
        # Annotations
        for i, imp in enumerate(feature_imp['importance']):
            ax.text(imp + 0.001, i, f'{imp:.3f}', va='center', fontweight='bold')
    
    # 3. Regionale Risk-Heatmap
    ax = axes[1, 0]
    
    if risk_scores and 'regional_risk' in risk_scores:
        risk_data = risk_scores['regional_risk'].sort_values('risk_score')
        
        colors = ['green' if score < 20 else 'yellow' if score < 40 else 'orange' if score < 60 else 'red' 
                 for score in risk_data['risk_score']]
        
        bars = ax.barh(range(len(risk_data)), risk_data['risk_score'], color=colors, alpha=0.7)
        ax.set_yticks(range(len(risk_data)))
        ax.set_yticklabels(risk_data['region'])
        ax.set_xlabel('Risk-Score (0-100)')
        ax.set_title('Regionale Risk-Scores', fontweight='bold')
        ax.grid(True, alpha=0.3)
        
        # Annotations
        for i, score in enumerate(risk_data['risk_score']):
            ax.text(score + 1, i, f'{score:.1f}', va='center', fontweight='bold')
    
    # 4. Zeitliche Anomalie-Patterns
    ax = axes[1, 1]
    
    # Anomalie-Rate nach Tageszeit
    hourly_anomalies = features_df.groupby('hour').agg({
        'is_anomaly': ['mean', 'count']
    }).reset_index()
    hourly_anomalies.columns = ['hour', 'anomaly_rate', 'count']
    
    # Nur Stunden mit gen√ºgend Daten
    hourly_anomalies = hourly_anomalies[hourly_anomalies['count'] > 100]
    
    if len(hourly_anomalies) > 0:
        ax.plot(hourly_anomalies['hour'], hourly_anomalies['anomaly_rate'] * 100, 
               marker='o', linewidth=2, markersize=6)
        ax.set_xlabel('Stunde (UTC)')
        ax.set_ylabel('Anomalie-Rate (%)')
        ax.set_title('Anomalie-Rate nach Tageszeit', fontweight='bold')
        ax.grid(True, alpha=0.3)
        ax.set_xticks(range(0, 24, 3))
        
        # Highlight Peak-Stunden
        peak_hour = hourly_anomalies.loc[hourly_anomalies['anomaly_rate'].idxmax(), 'hour']
        ax.axvline(x=peak_hour, color='red', linestyle='--', alpha=0.7, label=f'Peak: {peak_hour}:00')
        ax.legend()
    
    # 5. Provider-Performance vs. Anomalie-Rate
    ax = axes[2, 0]
    
    provider_stats = features_df.groupby('provider').agg({
        'final_latency': 'mean',
        'is_anomaly': 'mean'
    }).reset_index()
    
    # Nur Provider mit gen√ºgend Daten
    provider_counts = features_df['provider'].value_counts()
    provider_stats = provider_stats[provider_stats['provider'].isin(provider_counts[provider_counts > 1000].index)]
    
    if len(provider_stats) > 0:
        colors = ['red' if 'Akamai' in provider else 'lightblue' for provider in provider_stats['provider']]
        
        scatter = ax.scatter(provider_stats['final_latency'], provider_stats['is_anomaly'] * 100, 
                           s=200, c=colors, alpha=0.7, edgecolors='black', linewidths=2)
        
        ax.set_xlabel('Durchschnittliche Latenz (ms)')
        ax.set_ylabel('Anomalie-Rate (%)')
        ax.set_title('Provider-Performance vs. Anomalie-Rate', fontweight='bold')
        ax.grid(True, alpha=0.3)
        
        # Annotations
        for _, provider in provider_stats.iterrows():
            ax.annotate(provider['provider'], 
                       (provider['final_latency'], provider['is_anomaly'] * 100),
                       xytext=(5, 5), textcoords='offset points', fontweight='bold')
    
    # 6. ML-Modell-Performance-Vergleich
    ax = axes[2, 1]
    
    if ml_models and 'model_performance' in ml_models:
        model_perf = ml_models['model_performance'].dropna()
        
        if len(model_perf) > 0:
            x = np.arange(len(model_perf))
            width = 0.25
            
            bars1 = ax.bar(x - width, model_perf['Accuracy'], width, label='Accuracy', alpha=0.8)
            bars2 = ax.bar(x, model_perf['Precision'], width, label='Precision', alpha=0.8)
            bars3 = ax.bar(x + width, model_perf['Recall'], width, label='Recall', alpha=0.8)
            
            ax.set_xlabel('ML-Modelle')
            ax.set_ylabel('Performance-Metriken')
            ax.set_title('ML-Modell-Performance-Vergleich', fontweight='bold')
            ax.set_xticks(x)
            ax.set_xticklabels(model_perf['Model'], rotation=45, ha='right')
            ax.legend()
            ax.grid(True, alpha=0.3)
            ax.set_ylim(0, 1)
            
            # Annotations
            for i, (acc, prec, rec) in enumerate(zip(model_perf['Accuracy'], model_perf['Precision'], model_perf['Recall'])):
                if not pd.isna(acc):
                    ax.text(i - width, acc + 0.02, f'{acc:.2f}', ha='center', va='bottom', fontweight='bold')
                if not pd.isna(prec):
                    ax.text(i, prec + 0.02, f'{prec:.2f}', ha='center', va='bottom', fontweight='bold')
                if not pd.isna(rec):
                    ax.text(i + width, rec + 0.02, f'{rec:.2f}', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.show()

# ================================================================
# 7. HAUPTANALYSE-FUNKTION
# ================================================================

def run_advanced_anomaly_prediction_analysis():
    """F√ºhrt die komplette Advanced Anomalie-Vorhersage-Analyse durch"""
    
    # Dateipfade (anpassen falls n√∂tig)
    IPv4_FILE = "../data/IPv4.parquet"
    IPv6_FILE = "../data/IPv6.parquet"
    
    print("üö® STARTE ADVANCED ANOMALIE-VORHERSAGE DEEP-DIVE...")
    print("="*75)
    
    try:
        # Daten laden
        print("üìÇ Lade Daten...")
        df_ipv4 = pd.read_parquet(IPv4_FILE)
        df_ipv6 = pd.read_parquet(IPv6_FILE)
        print(f"‚úÖ IPv4: {df_ipv4.shape[0]:,} Messungen geladen")
        print(f"‚úÖ IPv6: {df_ipv6.shape[0]:,} Messungen geladen")
        
        # Analysen f√ºr beide Protokolle
        for protocol, df in [("IPv4", df_ipv4), ("IPv6", df_ipv6)]:
            print(f"\n{'='*75}")
            print(f"ADVANCED ANOMALIE-PREDICTION-ANALYSE F√úR {protocol}")
            print(f"{'='*75}")
            
            # 1. Feature-Extraktion
            features_df = extract_advanced_anomaly_features(df, protocol)
            
            if features_df is not None:
                # 2. Time-Series-Anomalie-Detection
                ts_results = perform_time_series_anomaly_detection(features_df, protocol)
                
                # 3. ML Anomalie-Prediction-Modelle
                ml_models = build_anomaly_prediction_models(features_df, protocol)
                
                # 4. Geografisches Risk-Scoring
                risk_scores = build_geographic_risk_scoring(features_df, protocol)
                
                # 5. Real-Time-Pipeline-Design
                pipeline = design_realtime_anomaly_pipeline(ml_models, risk_scores, protocol)
                
                # 6. Visualisierungen
                create_anomaly_prediction_visualizations(features_df, ml_models, risk_scores, protocol)
        
        print(f"\n{'='*75}")
        print("üéØ ADVANCED ANOMALIE-VORHERSAGE DEEP-DIVE ABGESCHLOSSEN!")
        print("ü§ñ PR√ÑDIKTIVE INTELLIGENCE-SYSTEME ENTWICKELT!")
        print("="*75)
        
        print(f"\nüîç WICHTIGSTE ERKENNTNISSE:")
        key_insights = [
            "‚úÖ Time-Series-basierte Anomalie-Detection implementiert",
            "‚úÖ Multi-Modell ML-Ensemble f√ºr Anomalie-Prediction entwickelt",
            "‚úÖ Geografisches Risk-Scoring-System etabliert",
            "‚úÖ Real-Time-Detection-Pipeline entworfen",
            "‚úÖ Predictive Failure-Detection-Algorithmen erstellt",
            "‚úÖ Provider-spezifische Risk-Assessment-Metriken",
            "‚úÖ Kontinentale Anomalie-Pattern identifiziert"
        ]
        
        for insight in key_insights:
            print(insight)
        
        print(f"\nüöÄ BEREIT F√úR:")
        print("  ‚Ä¢ Phase 4B3: Hop-Effizienz-Optimierung")
        print("  ‚Ä¢ Real-Time-Monitoring-Dashboard-Implementierung")
        print("  ‚Ä¢ Pr√§diktive Maintenance-Systeme")
        
    except Exception as e:
        print(f"‚ùå Fehler in der Anomalie-Prediction-Analyse: {e}")
        import traceback
        traceback.print_exc()

# Hauptanalyse ausf√ºhren
if __name__ == "__main__":
    run_advanced_anomaly_prediction_analysis()
```
