```python
# Phase 4: Umfassende Visualisierungen - MTR Anycast Studie
# ================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Für erweiterte Visualisierungen (optional)
try:
    import plotly.express as px
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    import plotly.figure_factory as ff
    PLOTLY_AVAILABLE = True
    print("✅ Plotly verfügbar - interaktive Visualisierungen aktiviert")
except ImportError:
    PLOTLY_AVAILABLE = False
    print("⚠️ Plotly nicht verfügbar - nur statische Visualisierungen")

plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (15, 10)

print("=== PHASE 4: UMFASSENDE VISUALISIERUNGEN ===")
print("Visuelle Darstellung aller erweiterten Analysen")
print("="*50)

# ================================================================
# 1. DATEN VORBEREITEN FÜR VISUALISIERUNGEN
# ================================================================

def prepare_visualization_data():
    """Bereitet alle Daten für Visualisierungen vor (basierend auf Phase 4 Ergebnissen)"""
    
    # Simuliere Daten basierend auf Phase 4 Ergebnissen
    # (In der Praxis würden diese aus den tatsächlichen Analysen kommen)
    
    print("\n1. DATEN VORBEREITEN FÜR VISUALISIERUNGEN...")
    print("-" * 50)
    
    # AWS Regionen
    regions = ['eu-central-1', 'eu-north-1', 'us-west-1', 'ca-central-1', 
               'af-south-1', 'ap-east-1', 'ap-south-1', 'ap-northeast-1', 
               'ap-southeast-2', 'sa-east-1']
    
    # Anomalie-Raten aus Phase 4 (IPv4)
    anomaly_rates_ipv4 = {
        'af-south-1': 33.43, 'eu-north-1': 29.20, 'ap-east-1': 22.42,
        'us-west-1': 20.52, 'ca-central-1': 13.68, 'ap-southeast-2': 12.15,
        'sa-east-1': 7.04, 'ap-south-1': 5.38, 'ap-northeast-1': 3.74,
        'eu-central-1': 3.34
    }
    
    # Anomalie-Raten IPv6
    anomaly_rates_ipv6 = {
        'af-south-1': 50.64, 'eu-north-1': 27.30, 'ca-central-1': 26.73,
        'ap-east-1': 20.42, 'ap-south-1': 20.18, 'us-west-1': 17.04,
        'sa-east-1': 13.38, 'eu-central-1': 7.17, 'ap-southeast-2': 4.92,
        'ap-northeast-1': 3.94
    }
    
    # Provider-Performance aus Phase 4
    provider_performance = {
        'IPv4': {
            'Cloudflare': {'latency': 1.82, 'sla_score': 99.9, 'edge_density': 2.0, 'anomaly_rate': 13.96},
            'Google': {'latency': 3.69, 'sla_score': 96.0, 'edge_density': 1.0, 'anomaly_rate': 14.98},
            'Quad9': {'latency': 2.85, 'sla_score': 95.8, 'edge_density': 1.0, 'anomaly_rate': 15.49},
            'Akamai': {'latency': 145.59, 'sla_score': 50.0, 'edge_density': 1.1, 'anomaly_rate': 20.38}
        },
        'IPv6': {
            'Cloudflare': {'latency': 2.04, 'sla_score': 99.9, 'edge_density': 2.0, 'anomaly_rate': 16.77},
            'Google': {'latency': 5.60, 'sla_score': 94.8, 'edge_density': 1.0, 'anomaly_rate': 29.52},
            'Quad9': {'latency': 3.09, 'sla_score': 95.9, 'edge_density': 1.0, 'anomaly_rate': 18.99},
            'Akamai': {'latency': 144.65, 'sla_score': 52.0, 'edge_density': 1.1, 'anomaly_rate': 16.67}
        }
    }
    
    # Hop-Count-Daten aus Phase 4
    hop_counts = {
        'IPv4': {
            'Anycast': {'mean': 7.6, 'std': 2.0, 'min': 2, 'max': 18},
            'Pseudo-Anycast': {'mean': 18.6, 'std': 3.5, 'min': 12, 'max': 30},
            'Unicast': {'mean': 16.9, 'std': 4.6, 'min': 5, 'max': 27}
        },
        'IPv6': {
            'Anycast': {'mean': 9.1, 'std': 2.4, 'min': 4, 'max': 19},
            'Pseudo-Anycast': {'mean': 16.8, 'std': 3.7, 'min': 8, 'max': 25},
            'Unicast': {'mean': 17.6, 'std': 5.1, 'min': 6, 'max': 30}
        }
    }
    
    # Korrelations-Daten aus Phase 4
    correlations_ipv4 = {
        'latency_hops': 0.801,
        'latency_failures': 0.375,
        'hops_failures': 0.639
    }
    
    correlations_ipv6 = {
        'latency_hops': 0.732,
        'latency_failures': 0.366,
        'hops_failures': 0.716
    }
    
    # Anomalie-Typen aus Phase 4
    anomaly_types_ipv4 = {
        'latency_spike': 8798, 'jitter_spike': 7980, 
        'routing_change': 7425, 'packet_loss': 87
    }
    
    anomaly_types_ipv6 = {
        'routing_change': 13695, 'latency_spike': 9094,
        'jitter_spike': 8043, 'packet_loss': 31
    }
    
    # Feature-Wichtigkeit aus Machine Learning
    feature_importance = {
        'IPv4': {'service_type': 0.662, 'total_hops': 0.315, 'hour': 0.013, 'day_of_week': 0.009},
        'IPv6': {'service_type': 0.785, 'total_hops': 0.198, 'hour': 0.010, 'day_of_week': 0.007}
    }
    
    return {
        'regions': regions,
        'anomaly_rates_ipv4': anomaly_rates_ipv4,
        'anomaly_rates_ipv6': anomaly_rates_ipv6,
        'provider_performance': provider_performance,
        'hop_counts': hop_counts,
        'correlations_ipv4': correlations_ipv4,
        'correlations_ipv6': correlations_ipv6,
        'anomaly_types_ipv4': anomaly_types_ipv4,
        'anomaly_types_ipv6': anomaly_types_ipv6,
        'feature_importance': feature_importance
    }

# Daten vorbereiten
viz_data = prepare_visualization_data()

# ================================================================
# 2. GEOGRAFISCHE ANOMALIE-HEATMAPS
# ================================================================

def create_geographic_anomaly_heatmaps(viz_data):
    """Erstellt geografische Anomalie-Heatmaps"""
    print("\n2. GEOGRAFISCHE ANOMALIE-HEATMAPS")
    print("-" * 40)
    
    fig, axes = plt.subplots(1, 2, figsize=(20, 8))
    fig.suptitle('Geografische Anomalie-Verteilung: Das Afrika-Problem', fontsize=16, fontweight='bold')
    
    # IPv4 Anomalie-Heatmap
    regions = list(viz_data['anomaly_rates_ipv4'].keys())
    rates_ipv4 = list(viz_data['anomaly_rates_ipv4'].values())
    
    colors_ipv4 = plt.cm.Reds(np.linspace(0.3, 1.0, len(regions)))
    bars1 = axes[0].barh(regions, rates_ipv4, color=colors_ipv4)
    axes[0].set_title('IPv4 Anomalie-Raten pro Region', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Anomalie-Rate (%)')
    axes[0].grid(axis='x', alpha=0.3)
    
    # Annotations hinzufügen
    for i, (region, rate) in enumerate(zip(regions, rates_ipv4)):
        axes[0].text(rate + 1, i, f'{rate:.1f}%', va='center', fontweight='bold')
    
    # Highlight extreme Werte
    axes[0].axvline(x=30, color='red', linestyle='--', alpha=0.7, label='Kritische Schwelle')
    axes[0].legend()
    
    # IPv6 Anomalie-Heatmap
    regions_ipv6 = list(viz_data['anomaly_rates_ipv6'].keys())
    rates_ipv6 = list(viz_data['anomaly_rates_ipv6'].values())
    
    colors_ipv6 = plt.cm.Reds(np.linspace(0.3, 1.0, len(regions_ipv6)))
    bars2 = axes[1].barh(regions_ipv6, rates_ipv6, color=colors_ipv6)
    axes[1].set_title('IPv6 Anomalie-Raten pro Region', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Anomalie-Rate (%)')
    axes[1].grid(axis='x', alpha=0.3)
    
    # Annotations
    for i, (region, rate) in enumerate(zip(regions_ipv6, rates_ipv6)):
        axes[1].text(rate + 1, i, f'{rate:.1f}%', va='center', fontweight='bold')
    
    axes[1].axvline(x=30, color='red', linestyle='--', alpha=0.7, label='Kritische Schwelle')
    axes[1].legend()
    
    plt.tight_layout()
    plt.show()
    
    # Zusätzliche Vergleichsvisualisierung
    fig, ax = plt.subplots(figsize=(15, 8))
    
    x = np.arange(len(regions))
    width = 0.35
    
    # Sortiere für bessere Darstellung
    combined_data = [(r, viz_data['anomaly_rates_ipv4'][r], viz_data['anomaly_rates_ipv6'][r]) 
                     for r in regions if r in viz_data['anomaly_rates_ipv6']]
    combined_data.sort(key=lambda x: x[1], reverse=True)
    
    regions_sorted = [x[0] for x in combined_data]
    ipv4_sorted = [x[1] for x in combined_data]
    ipv6_sorted = [x[2] for x in combined_data]
    
    bars1 = ax.bar(x - width/2, ipv4_sorted, width, label='IPv4', alpha=0.8, color='skyblue')
    bars2 = ax.bar(x + width/2, ipv6_sorted, width, label='IPv6', alpha=0.8, color='lightcoral')
    
    ax.set_title('IPv4 vs IPv6 Anomalie-Raten: Protokoll-Vergleich', fontsize=16, fontweight='bold')
    ax.set_xlabel('AWS Regionen')
    ax.set_ylabel('Anomalie-Rate (%)')
    ax.set_xticks(x)
    ax.set_xticklabels(regions_sorted, rotation=45, ha='right')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    
    # Highlight Afrika-Problem
    africa_index = regions_sorted.index('af-south-1') if 'af-south-1' in regions_sorted else -1
    if africa_index >= 0:
        ax.annotate('🚨 Afrika-Problem!\n15x höher als beste Region', 
                   xy=(africa_index, max(ipv4_sorted[africa_index], ipv6_sorted[africa_index])),
                   xytext=(africa_index + 2, max(ipv4_sorted[africa_index], ipv6_sorted[africa_index]) + 10),
                   arrowprops=dict(arrowstyle='->', color='red', lw=2),
                   fontsize=12, fontweight='bold', color='red',
                   bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
    
    plt.tight_layout()
    plt.show()

# ================================================================
# 3. PROVIDER-INFRASTRUKTUR-DASHBOARD
# ================================================================

def create_provider_infrastructure_dashboard(viz_data):
    """Erstellt Provider-Infrastruktur-Dashboard"""
    print("\n3. PROVIDER-INFRASTRUKTUR-DASHBOARD")
    print("-" * 40)
    
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    fig.suptitle('Provider-Performance-Dashboard: Cloudflare dominiert', fontsize=16, fontweight='bold')
    
    protocols = ['IPv4', 'IPv6']
    providers = ['Cloudflare', 'Google', 'Quad9', 'Akamai']
    
    # 1. Latenz-Vergleich
    for i, protocol in enumerate(protocols):
        latencies = [viz_data['provider_performance'][protocol][p]['latency'] for p in providers]
        colors = ['gold', 'lightgreen', 'lightblue', 'lightcoral']
        
        bars = axes[0, i].bar(providers, latencies, color=colors, alpha=0.8)
        axes[0, i].set_title(f'{protocol} Latenz-Performance', fontsize=12, fontweight='bold')
        axes[0, i].set_ylabel('Latenz (ms)')
        axes[0, i].set_yscale('log')  # Log-Scale wegen Akamai-Ausreißer
        
        # Annotations
        for bar, latency in zip(bars, latencies):
            height = bar.get_height()
            axes[0, i].text(bar.get_x() + bar.get_width()/2., height + height*0.1,
                           f'{latency:.1f}ms', ha='center', va='bottom', fontweight='bold')
        
        axes[0, i].tick_params(axis='x', rotation=45)
        axes[0, i].grid(axis='y', alpha=0.3)
    
    # 2. SLA-Score-Vergleich
    for i, protocol in enumerate(protocols):
        sla_scores = [viz_data['provider_performance'][protocol][p]['sla_score'] for p in providers]
        
        bars = axes[1, i].bar(providers, sla_scores, color=colors, alpha=0.8)
        axes[1, i].set_title(f'{protocol} SLA-Scores', fontsize=12, fontweight='bold')
        axes[1, i].set_ylabel('SLA-Score (/100)')
        axes[1, i].set_ylim(0, 100)
        
        for bar, score in zip(bars, sla_scores):
            height = bar.get_height()
            axes[1, i].text(bar.get_x() + bar.get_width()/2., height + 2,
                           f'{score:.1f}', ha='center', va='bottom', fontweight='bold')
        
        axes[1, i].tick_params(axis='x', rotation=45)
        axes[1, i].grid(axis='y', alpha=0.3)
    
    # 3. Provider-Effizienz-Matrix (Kombiniert)
    # Kombiniere IPv4 und IPv6 Durchschnittswerte
    combined_latency = []
    combined_sla = []
    combined_anomaly = []
    
    for provider in providers:
        avg_latency = np.mean([viz_data['provider_performance'][p][provider]['latency'] for p in protocols])
        avg_sla = np.mean([viz_data['provider_performance'][p][provider]['sla_score'] for p in protocols])
        avg_anomaly = np.mean([viz_data['provider_performance'][p][provider]['anomaly_rate'] for p in protocols])
        
        combined_latency.append(avg_latency)
        combined_sla.append(avg_sla)
        combined_anomaly.append(avg_anomaly)
    
    # Scatter plot: SLA vs Latenz (mit Anomalie-Rate als Größe)
    ax = axes[0, 2]
    scatter = ax.scatter(combined_latency, combined_sla, 
                        s=[rate*10 for rate in combined_anomaly], 
                        c=colors, alpha=0.7, edgecolors='black', linewidths=2)
    
    ax.set_xlabel('Durchschnittliche Latenz (ms)')
    ax.set_ylabel('Durchschnittlicher SLA-Score')
    ax.set_title('Provider-Effizienz-Matrix\n(Blasengröße = Anomalie-Rate)', fontsize=12, fontweight='bold')
    ax.set_xscale('log')
    ax.grid(True, alpha=0.3)
    
    # Annotations
    for i, provider in enumerate(providers):
        ax.annotate(provider, (combined_latency[i], combined_sla[i]), 
                   xytext=(5, 5), textcoords='offset points', fontweight='bold')
    
    # 4. Anomalie-Rate-Vergleich
    ax = axes[1, 2]
    
    ipv4_anomalies = [viz_data['provider_performance']['IPv4'][p]['anomaly_rate'] for p in providers]
    ipv6_anomalies = [viz_data['provider_performance']['IPv6'][p]['anomaly_rate'] for p in providers]
    
    x = np.arange(len(providers))
    width = 0.35
    
    bars1 = ax.bar(x - width/2, ipv4_anomalies, width, label='IPv4', alpha=0.8, color='skyblue')
    bars2 = ax.bar(x + width/2, ipv6_anomalies, width, label='IPv6', alpha=0.8, color='lightcoral')
    
    ax.set_title('Provider-Anomalie-Raten', fontsize=12, fontweight='bold')
    ax.set_ylabel('Anomalie-Rate (%)')
    ax.set_xticks(x)
    ax.set_xticklabels(providers, rotation=45)
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# ================================================================
# 4. HOP-COUNT-EFFIZIENZ-PLOTS
# ================================================================

def create_hop_efficiency_plots(viz_data):
    """Erstellt Hop-Count-Effizienz-Visualisierungen"""
    print("\n4. HOP-COUNT-EFFIZIENZ-PLOTS")
    print("-" * 40)
    
    fig, axes = plt.subplots(2, 2, figsize=(18, 12))
    fig.suptitle('Hop-Count-Analyse: Anycast-Effizienz vs. Akamai-Problem', fontsize=16, fontweight='bold')
    
    service_types = ['Anycast', 'Pseudo-Anycast', 'Unicast']
    colors = ['lightgreen', 'orange', 'lightcoral']
    
    # 1. IPv4 Hop-Count-Vergleich
    ax = axes[0, 0]
    hop_means_ipv4 = [viz_data['hop_counts']['IPv4'][st]['mean'] for st in service_types]
    hop_stds_ipv4 = [viz_data['hop_counts']['IPv4'][st]['std'] for st in service_types]
    
    bars = ax.bar(service_types, hop_means_ipv4, yerr=hop_stds_ipv4, 
                  color=colors, alpha=0.8, capsize=5, error_kw={'linewidth': 2})
    ax.set_title('IPv4 Hop-Count-Vergleich', fontsize=14, fontweight='bold')
    ax.set_ylabel('Durchschnittliche Hop-Anzahl')
    ax.grid(axis='y', alpha=0.3)
    
    for bar, mean, std in zip(bars, hop_means_ipv4, hop_stds_ipv4):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + std + 0.5,
               f'{mean:.1f}±{std:.1f}', ha='center', va='bottom', fontweight='bold')
    
    # Highlight Akamai-Problem
    ax.annotate('🚨 Akamai = Unicast!\nKein Anycast-Vorteil', 
               xy=(1, hop_means_ipv4[1]), xytext=(1.5, hop_means_ipv4[1] + 3),
               arrowprops=dict(arrowstyle='->', color='red', lw=2),
               fontsize=10, fontweight='bold', color='red',
               bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
    
    # 2. IPv6 Hop-Count-Vergleich
    ax = axes[0, 1]
    hop_means_ipv6 = [viz_data['hop_counts']['IPv6'][st]['mean'] for st in service_types]
    hop_stds_ipv6 = [viz_data['hop_counts']['IPv6'][st]['std'] for st in service_types]
    
    bars = ax.bar(service_types, hop_means_ipv6, yerr=hop_stds_ipv6, 
                  color=colors, alpha=0.8, capsize=5, error_kw={'linewidth': 2})
    ax.set_title('IPv6 Hop-Count-Vergleich', fontsize=14, fontweight='bold')
    ax.set_ylabel('Durchschnittliche Hop-Anzahl')
    ax.grid(axis='y', alpha=0.3)
    
    for bar, mean, std in zip(bars, hop_means_ipv6, hop_stds_ipv6):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + std + 0.5,
               f'{mean:.1f}±{std:.1f}', ha='center', va='bottom', fontweight='bold')
    
    # 3. Protokoll-Vergleich
    ax = axes[1, 0]
    x = np.arange(len(service_types))
    width = 0.35
    
    bars1 = ax.bar(x - width/2, hop_means_ipv4, width, label='IPv4', 
                   color='skyblue', alpha=0.8)
    bars2 = ax.bar(x + width/2, hop_means_ipv6, width, label='IPv6', 
                   color='lightcoral', alpha=0.8)
    
    ax.set_title('IPv4 vs IPv6 Hop-Count-Vergleich', fontsize=14, fontweight='bold')
    ax.set_ylabel('Durchschnittliche Hop-Anzahl')
    ax.set_xticks(x)
    ax.set_xticklabels(service_types)
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    
    # 4. Effizienz-Score (Unicast als Baseline)
    ax = axes[1, 1]
    
    # Berechne Effizienz als Ratio zu Unicast
    efficiency_ipv4 = [viz_data['hop_counts']['IPv4']['Unicast']['mean'] / viz_data['hop_counts']['IPv4'][st]['mean'] 
                      for st in service_types]
    efficiency_ipv6 = [viz_data['hop_counts']['IPv6']['Unicast']['mean'] / viz_data['hop_counts']['IPv6'][st]['mean'] 
                      for st in service_types]
    
    bars1 = ax.bar(x - width/2, efficiency_ipv4, width, label='IPv4', 
                   color='skyblue', alpha=0.8)
    bars2 = ax.bar(x + width/2, efficiency_ipv6, width, label='IPv6', 
                   color='lightcoral', alpha=0.8)
    
    ax.set_title('Hop-Effizienz-Score (vs. Unicast)', fontsize=14, fontweight='bold')
    ax.set_ylabel('Effizienz-Faktor')
    ax.set_xticks(x)
    ax.set_xticklabels(service_types)
    ax.axhline(y=1, color='red', linestyle='--', alpha=0.7, label='Unicast-Baseline')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    
    # Annotations für Effizienz
    for i, (eff4, eff6) in enumerate(zip(efficiency_ipv4, efficiency_ipv6)):
        ax.text(i - width/2, eff4 + 0.05, f'{eff4:.1f}x', ha='center', va='bottom', fontweight='bold')
        ax.text(i + width/2, eff6 + 0.05, f'{eff6:.1f}x', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.show()

# ================================================================
# 5. KORRELATIONS-MATRIX-HEATMAPS
# ================================================================

def create_correlation_heatmaps(viz_data):
    """Erstellt Korrelations-Matrix-Heatmaps"""
    print("\n5. KORRELATIONS-MATRIX-HEATMAPS")
    print("-" * 40)
    
    # Erstelle Korrelationsmatrizen basierend auf Phase 4 Daten
    corr_matrix_ipv4 = np.array([
        [1.000, 0.027, 0.057, 0.375, 0.801],
        [0.027, 1.000, 0.019, -0.014, -0.009],
        [0.057, 0.019, 1.000, 0.001, 0.015],
        [0.375, -0.014, 0.001, 1.000, 0.639],
        [0.801, -0.009, 0.015, 0.639, 1.000]
    ])
    
    corr_matrix_ipv6 = np.array([
        [1.000, 0.034, 0.089, 0.366, 0.732],
        [0.034, 1.000, 0.034, 0.010, 0.015],
        [0.089, 0.034, 1.000, 0.044, 0.051],
        [0.366, 0.010, 0.044, 1.000, 0.716],
        [0.732, 0.015, 0.051, 0.716, 1.000]
    ])
    
    labels = ['Latenz', 'Packet Loss', 'Jitter', 'Failures', 'Total Hops']
    
    fig, axes = plt.subplots(1, 2, figsize=(18, 8))
    fig.suptitle('Korrelations-Matrix: Starke Latenz-Hop-Korrelation entdeckt', fontsize=16, fontweight='bold')
    
    # IPv4 Korrelationen
    im1 = axes[0].imshow(corr_matrix_ipv4, cmap='RdBu_r', vmin=-1, vmax=1)
    axes[0].set_title('IPv4 Performance-Korrelationen', fontsize=14, fontweight='bold')
    axes[0].set_xticks(range(len(labels)))
    axes[0].set_yticks(range(len(labels)))
    axes[0].set_xticklabels(labels, rotation=45, ha='right')
    axes[0].set_yticklabels(labels)
    
    # Annotations hinzufügen
    for i in range(len(labels)):
        for j in range(len(labels)):
            text = f'{corr_matrix_ipv4[i, j]:.3f}'
            color = 'white' if abs(corr_matrix_ipv4[i, j]) > 0.5 else 'black'
            weight = 'bold' if abs(corr_matrix_ipv4[i, j]) > 0.3 else 'normal'
            axes[0].text(j, i, text, ha='center', va='center', 
                        color=color, fontweight=weight)
    
    # IPv6 Korrelationen
    im2 = axes[1].imshow(corr_matrix_ipv6, cmap='RdBu_r', vmin=-1, vmax=1)
    axes[1].set_title('IPv6 Performance-Korrelationen', fontsize=14, fontweight='bold')
    axes[1].set_xticks(range(len(labels)))
    axes[1].set_yticks(range(len(labels)))
    axes[1].set_xticklabels(labels, rotation=45, ha='right')
    axes[1].set_yticklabels(labels)
    
    for i in range(len(labels)):
        for j in range(len(labels)):
            text = f'{corr_matrix_ipv6[i, j]:.3f}'
            color = 'white' if abs(corr_matrix_ipv6[i, j]) > 0.5 else 'black'
            weight = 'bold' if abs(corr_matrix_ipv6[i, j]) > 0.3 else 'normal'
            axes[1].text(j, i, text, ha='center', va='center', 
                        color=color, fontweight=weight)
    
    # Colorbar
    cbar = plt.colorbar(im2, ax=axes, orientation='horizontal', pad=0.1, shrink=0.8)
    cbar.set_label('Korrelationskoeffizient', fontsize=12, fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    # Zusätzliche starke Korrelationen hervorheben
    fig, ax = plt.subplots(figsize=(12, 8))
    
    correlations = ['Latenz ↔ Hops', 'Hops ↔ Failures', 'Latenz ↔ Failures']
    ipv4_values = [0.801, 0.639, 0.375]
    ipv6_values = [0.732, 0.716, 0.366]
    
    x = np.arange(len(correlations))
    width = 0.35
    
    bars1 = ax.bar(x - width/2, ipv4_values, width, label='IPv4', 
                   color='skyblue', alpha=0.8)
    bars2 = ax.bar(x + width/2, ipv6_values, width, label='IPv6', 
                   color='lightcoral', alpha=0.8)
    
    ax.set_title('Starke Korrelationen: Netzwerk-Performance-Zusammenhänge', 
                 fontsize=16, fontweight='bold')
    ax.set_ylabel('Korrelationskoeffizient')
    ax.set_xticks(x)
    ax.set_xticklabels(correlations)
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    ax.axhline(y=0.3, color='orange', linestyle='--', alpha=0.7, label='Signifikanz-Schwelle')
    ax.axhline(y=0.7, color='red', linestyle='--', alpha=0.7, label='Starke Korrelation')
    
    # Annotations
    for i, (v4, v6) in enumerate(zip(ipv4_values, ipv6_values)):
        ax.text(i - width/2, v4 + 0.02, f'{v4:.3f}', ha='center', va='bottom', fontweight='bold')
        ax.text(i + width/2, v6 + 0.02, f'{v6:.3f}', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.show()

# ================================================================
# 6. ANOMALIE-TYPEN-VERTEILUNGEN
# ================================================================

def create_anomaly_type_distributions(viz_data):
    """Erstellt Anomalie-Typen-Verteilungsplots"""
    print("\n6. ANOMALIE-TYPEN-VERTEILUNGEN")
    print("-" * 40)
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Anomalie-Typen-Analyse: Routing-Changes dominieren IPv6', fontsize=16, fontweight='bold')
    
    # IPv4 Anomalie-Typen (Pie Chart)
    ax = axes[0, 0]
    labels_ipv4 = list(viz_data['anomaly_types_ipv4'].keys())
    sizes_ipv4 = list(viz_data['anomaly_types_ipv4'].values())
    colors = ['lightcoral', 'lightskyblue', 'lightgreen', 'gold']
    
    wedges, texts, autotexts = ax.pie(sizes_ipv4, labels=labels_ipv4, colors=colors, 
                                     autopct='%1.1f%%', startangle=90, 
                                     textprops={'fontweight': 'bold'})
    ax.set_title('IPv4 Anomalie-Typen-Verteilung', fontsize=14, fontweight='bold')
    
    # IPv6 Anomalie-Typen (Pie Chart)
    ax = axes[0, 1]
    labels_ipv6 = list(viz_data['anomaly_types_ipv6'].keys())
    sizes_ipv6 = list(viz_data['anomaly_types_ipv6'].values())
    
    wedges, texts, autotexts = ax.pie(sizes_ipv6, labels=labels_ipv6, colors=colors, 
                                     autopct='%1.1f%%', startangle=90, 
                                     textprops={'fontweight': 'bold'})
    ax.set_title('IPv6 Anomalie-Typen-Verteilung', fontsize=14, fontweight='bold')
    
    # Vergleich der Anomalie-Typen
    ax = axes[1, 0]
    
    # Normalisiere für Vergleichbarkeit
    total_ipv4 = sum(sizes_ipv4)
    total_ipv6 = sum(sizes_ipv6)
    
    normalized_ipv4 = [s/total_ipv4*100 for s in sizes_ipv4]
    normalized_ipv6 = [s/total_ipv6*100 for s in sizes_ipv6]
    
    x = np.arange(len(labels_ipv4))
    width = 0.35
    
    bars1 = ax.bar(x - width/2, normalized_ipv4, width, label='IPv4', 
                   color='skyblue', alpha=0.8)
    bars2 = ax.bar(x + width/2, normalized_ipv6, width, label='IPv6', 
                   color='lightcoral', alpha=0.8)
    
    ax.set_title('Anomalie-Typen: IPv4 vs IPv6 Vergleich', fontsize=14, fontweight='bold')
    ax.set_ylabel('Anteil (%)')
    ax.set_xticks(x)
    ax.set_xticklabels(labels_ipv4, rotation=45, ha='right')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    
    # Annotations
    for i, (v4, v6) in enumerate(zip(normalized_ipv4, normalized_ipv6)):
        ax.text(i - width/2, v4 + 1, f'{v4:.1f}%', ha='center', va='bottom', fontweight='bold')
        ax.text(i + width/2, v6 + 1, f'{v6:.1f}%', ha='center', va='bottom', fontweight='bold')
    
    # Absolute Zahlen-Vergleich
    ax = axes[1, 1]
    
    bars1 = ax.bar(x - width/2, sizes_ipv4, width, label='IPv4', 
                   color='skyblue', alpha=0.8)
    bars2 = ax.bar(x + width/2, sizes_ipv6, width, label='IPv6', 
                   color='lightcoral', alpha=0.8)
    
    ax.set_title('Absolute Anomalie-Zahlen', fontsize=14, fontweight='bold')
    ax.set_ylabel('Anzahl Anomalien')
    ax.set_xticks(x)
    ax.set_xticklabels(labels_ipv4, rotation=45, ha='right')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    ax.set_yscale('log')  # Log-scale für bessere Darstellung
    
    plt.tight_layout()
    plt.show()

# ================================================================
# 7. MACHINE LEARNING FEATURE-WICHTIGKEIT
# ================================================================

def create_ml_feature_importance_plots(viz_data):
    """Erstellt Feature-Wichtigkeit-Plots für Machine Learning Modelle"""
    print("\n7. MACHINE LEARNING FEATURE-WICHTIGKEIT")
    print("-" * 40)
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    fig.suptitle('Machine Learning Insights: Service-Typ erklärt 66-78% der Performance', 
                 fontsize=16, fontweight='bold')
    
    # IPv4 Feature-Wichtigkeit
    ax = axes[0, 0]
    features_ipv4 = list(viz_data['feature_importance']['IPv4'].keys())
    importance_ipv4 = list(viz_data['feature_importance']['IPv4'].values())
    
    colors = ['gold', 'lightgreen', 'lightcoral', 'lightskyblue']
    bars = ax.barh(features_ipv4, importance_ipv4, color=colors, alpha=0.8)
    ax.set_title('IPv4 Feature-Wichtigkeit\n(R² = 0.863)', fontsize=14, fontweight='bold')
    ax.set_xlabel('Wichtigkeit')
    ax.grid(axis='x', alpha=0.3)
    
    for bar, imp in zip(bars, importance_ipv4):
        width = bar.get_width()
        ax.text(width + 0.01, bar.get_y() + bar.get_height()/2,
               f'{imp:.3f}', ha='left', va='center', fontweight='bold')
    
    # IPv6 Feature-Wichtigkeit
    ax = axes[0, 1]
    features_ipv6 = list(viz_data['feature_importance']['IPv6'].keys())
    importance_ipv6 = list(viz_data['feature_importance']['IPv6'].values())
    
    bars = ax.barh(features_ipv6, importance_ipv6, color=colors, alpha=0.8)
    ax.set_title('IPv6 Feature-Wichtigkeit\n(R² = 0.827)', fontsize=14, fontweight='bold')
    ax.set_xlabel('Wichtigkeit')
    ax.grid(axis='x', alpha=0.3)
    
    for bar, imp in zip(bars, importance_ipv6):
        width = bar.get_width()
        ax.text(width + 0.01, bar.get_y() + bar.get_height()/2,
               f'{imp:.3f}', ha='left', va='center', fontweight='bold')
    
    # Vergleichsplot
    ax = axes[1, 0]
    x = np.arange(len(features_ipv4))
    width = 0.35
    
    bars1 = ax.bar(x - width/2, importance_ipv4, width, label='IPv4', 
                   color='skyblue', alpha=0.8)
    bars2 = ax.bar(x + width/2, importance_ipv6, width, label='IPv6', 
                   color='lightcoral', alpha=0.8)
    
    ax.set_title('Feature-Wichtigkeit: IPv4 vs IPv6', fontsize=14, fontweight='bold')
    ax.set_ylabel('Wichtigkeit')
    ax.set_xticks(x)
    ax.set_xticklabels(features_ipv4, rotation=45, ha='right')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    
    # Performance-Kategorien-Vorhersage (simuliert)
    ax = axes[1, 1]
    categories = ['Excellent', 'Good', 'Fair', 'Poor']
    # Basierend auf Phase 4 Daten: IPv4: 58.4%, 1.9%, 7.0%, 32.6%
    ipv4_categories = [58.4, 1.9, 7.0, 32.6]
    # IPv6: 58.1%, 2.0%, 8.0%, 32.0%
    ipv6_categories = [58.1, 2.0, 8.0, 32.0]
    
    x = np.arange(len(categories))
    bars1 = ax.bar(x - width/2, ipv4_categories, width, label='IPv4', 
                   color='skyblue', alpha=0.8)
    bars2 = ax.bar(x + width/2, ipv6_categories, width, label='IPv6', 
                   color='lightcoral', alpha=0.8)
    
    ax.set_title('Latenz-Kategorien-Vorhersage', fontsize=14, fontweight='bold')
    ax.set_ylabel('Anteil (%)')
    ax.set_xticks(x)
    ax.set_xticklabels(categories)
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# ================================================================
# 8. INTERAKTIVE PLOTLY-VISUALISIERUNG
# ================================================================

def create_interactive_dashboard(viz_data):
    """Erstellt interaktive Plotly-Visualisierung"""
    print("\n8. INTERAKTIVE PLOTLY-VISUALISIERUNG")
    print("-" * 40)
    
    try:
        import plotly.express as px
        import plotly.graph_objects as go
        from plotly.subplots import make_subplots
        
        # Bereite Daten für interaktive Plots vor
        regions = list(viz_data['anomaly_rates_ipv4'].keys())
        ipv4_rates = list(viz_data['anomaly_rates_ipv4'].values())
        ipv6_rates = [viz_data['anomaly_rates_ipv6'][r] for r in regions if r in viz_data['anomaly_rates_ipv6']]
        
        # Interaktive Anomalie-Heatmap
        fig = go.Figure()
        
        fig.add_trace(go.Bar(
            name='IPv4',
            x=regions,
            y=ipv4_rates,
            marker_color='lightblue',
            hovertemplate='<b>%{x}</b><br>IPv4 Anomalie-Rate: %{y:.1f}%<extra></extra>'
        ))
        
        fig.add_trace(go.Bar(
            name='IPv6',
            x=regions,
            y=ipv6_rates,
            marker_color='lightcoral',
            hovertemplate='<b>%{x}</b><br>IPv6 Anomalie-Rate: %{y:.1f}%<extra></extra>'
        ))
        
        fig.update_layout(
            title='Interaktive Anomalie-Raten pro Region',
            xaxis_title='AWS Regionen',
            yaxis_title='Anomalie-Rate (%)',
            barmode='group',
            hovermode='x unified',
            template='plotly_white'
        )
        
        # Sichere Alternative für show()
        try:
            fig.show()
            print("✅ Interaktive Plotly-Visualisierung erstellt")
        except Exception as show_error:
            print(f"⚠️ Plotly show() Fehler: {show_error}")
            print("💡 Versuche Alternative...")
            
            # Alternative: Speichere als HTML
            try:
                fig.write_html("anomaly_dashboard.html")
                print("✅ Interaktive Visualisierung als 'anomaly_dashboard.html' gespeichert")
                print("💡 Öffnen Sie die HTML-Datei im Browser für interaktive Ansicht")
            except Exception as html_error:
                print(f"⚠️ HTML-Export auch fehlgeschlagen: {html_error}")
                print("📊 Erstelle stattdessen statische Matplotlib-Alternative...")
                create_static_alternative_dashboard(viz_data)
        
    except ImportError as import_error:
        print(f"⚠️ Plotly Import-Fehler: {import_error}")
        print("💡 Install mit: pip install plotly nbformat")
        print("📊 Erstelle stattdessen statische Alternative...")
        create_static_alternative_dashboard(viz_data)
    except Exception as general_error:
        print(f"⚠️ Unerwarteter Plotly-Fehler: {general_error}")
        print("📊 Erstelle stattdessen statische Alternative...")
        create_static_alternative_dashboard(viz_data)

def create_static_alternative_dashboard(viz_data):
    """Erstellt eine statische Alternative zum interaktiven Dashboard"""
    print("\n📊 STATISCHE ALTERNATIVE ZUM INTERAKTIVEN DASHBOARD")
    print("-" * 50)
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Alternative Dashboard: Anomalie-Analyse (Statisch)', fontsize=16, fontweight='bold')
    
    # 1. Anomalie-Raten Vergleich
    regions = list(viz_data['anomaly_rates_ipv4'].keys())
    ipv4_rates = list(viz_data['anomaly_rates_ipv4'].values())
    ipv6_rates = [viz_data['anomaly_rates_ipv6'][r] for r in regions if r in viz_data['anomaly_rates_ipv6']]
    
    x = np.arange(len(regions))
    width = 0.35
    
    bars1 = axes[0, 0].bar(x - width/2, ipv4_rates, width, label='IPv4', 
                          color='skyblue', alpha=0.8)
    bars2 = axes[0, 0].bar(x + width/2, ipv6_rates, width, label='IPv6', 
                          color='lightcoral', alpha=0.8)
    
    axes[0, 0].set_title('Anomalie-Raten pro Region', fontsize=14, fontweight='bold')
    axes[0, 0].set_ylabel('Anomalie-Rate (%)')
    axes[0, 0].set_xticks(x)
    axes[0, 0].set_xticklabels(regions, rotation=45, ha='right')
    axes[0, 0].legend()
    axes[0, 0].grid(axis='y', alpha=0.3)
    
    # 2. Provider-Performance-Matrix
    providers = ['Cloudflare', 'Google', 'Quad9', 'Akamai']
    ipv4_latencies = [viz_data['provider_performance']['IPv4'][p]['latency'] for p in providers]
    ipv4_sla = [viz_data['provider_performance']['IPv4'][p]['sla_score'] for p in providers]
    
    scatter = axes[0, 1].scatter(ipv4_latencies, ipv4_sla, 
                                s=200, c=['gold', 'lightgreen', 'lightblue', 'lightcoral'], 
                                alpha=0.7, edgecolors='black', linewidths=2)
    
    axes[0, 1].set_xlabel('Latenz (ms)')
    axes[0, 1].set_ylabel('SLA-Score')
    axes[0, 1].set_title('Provider-Performance-Matrix', fontsize=14, fontweight='bold')
    axes[0, 1].set_xscale('log')
    axes[0, 1].grid(True, alpha=0.3)
    
    for i, provider in enumerate(providers):
        axes[0, 1].annotate(provider, (ipv4_latencies[i], ipv4_sla[i]), 
                           xytext=(5, 5), textcoords='offset points', fontweight='bold')
    
    # 3. Hop-Count-Effizienz
    service_types = ['Anycast', 'Pseudo-Anycast', 'Unicast']
    hop_means_ipv4 = [viz_data['hop_counts']['IPv4'][st]['mean'] for st in service_types]
    hop_means_ipv6 = [viz_data['hop_counts']['IPv6'][st]['mean'] for st in service_types]
    
    x_hops = np.arange(len(service_types))
    bars1 = axes[1, 0].bar(x_hops - width/2, hop_means_ipv4, width, label='IPv4', 
                          color='skyblue', alpha=0.8)
    bars2 = axes[1, 0].bar(x_hops + width/2, hop_means_ipv6, width, label='IPv6', 
                          color='lightcoral', alpha=0.8)
    
    axes[1, 0].set_title('Hop-Count-Vergleich', fontsize=14, fontweight='bold')
    axes[1, 0].set_ylabel('Durchschnittliche Hops')
    axes[1, 0].set_xticks(x_hops)
    axes[1, 0].set_xticklabels(service_types)
    axes[1, 0].legend()
    axes[1, 0].grid(axis='y', alpha=0.3)
    
    # 4. Feature-Wichtigkeit
    features = list(viz_data['feature_importance']['IPv4'].keys())
    importance_ipv4 = list(viz_data['feature_importance']['IPv4'].values())
    importance_ipv6 = list(viz_data['feature_importance']['IPv6'].values())
    
    x_feat = np.arange(len(features))
    bars1 = axes[1, 1].bar(x_feat - width/2, importance_ipv4, width, label='IPv4', 
                          color='skyblue', alpha=0.8)
    bars2 = axes[1, 1].bar(x_feat + width/2, importance_ipv6, width, label='IPv6', 
                          color='lightcoral', alpha=0.8)
    
    axes[1, 1].set_title('ML Feature-Wichtigkeit', fontsize=14, fontweight='bold')
    axes[1, 1].set_ylabel('Wichtigkeit')
    axes[1, 1].set_xticks(x_feat)
    axes[1, 1].set_xticklabels(features, rotation=45, ha='right')
    axes[1, 1].legend()
    axes[1, 1].grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("✅ Statische Alternative-Dashboard erstellt")

# ================================================================
# 9. HAUPT-VISUALISIERUNGS-FUNKTION
# ================================================================

def create_all_visualizations():
    """Erstellt alle Phase 4 Visualisierungen"""
    
    print("🎨 STARTE ALLE PHASE 4 VISUALISIERUNGEN...")
    print("="*50)
    
    # Daten vorbereiten
    viz_data = prepare_visualization_data()
    
    # Alle Visualisierungen erstellen
    create_geographic_anomaly_heatmaps(viz_data)
    create_provider_infrastructure_dashboard(viz_data)
    create_hop_efficiency_plots(viz_data)
    create_correlation_heatmaps(viz_data)
    create_anomaly_type_distributions(viz_data)
    create_ml_feature_importance_plots(viz_data)
    create_interactive_dashboard(viz_data)
    
    print("\n" + "="*60)
    print("🎯 ALLE PHASE 4 VISUALISIERUNGEN ABGESCHLOSSEN!")
    print("🏆 PUBLIKATIONSREIFE CHARTS ERSTELLT!")
    print("="*60)
    
    print("\n📊 ERSTELLTE VISUALISIERUNGEN:")
    charts = [
        "✅ Geografische Anomalie-Heatmaps (Das Afrika-Problem)",
        "✅ Provider-Infrastruktur-Dashboard (Cloudflare dominiert)", 
        "✅ Hop-Count-Effizienz-Plots (Anycast vs Akamai vs Unicast)",
        "✅ Korrelations-Matrix-Heatmaps (Starke Latenz-Hop-Korrelation)",
        "✅ Anomalie-Typen-Verteilungen (IPv4 vs IPv6 Unterschiede)",
        "✅ Machine Learning Feature-Wichtigkeit (Service-Typ = 66-78%)",
        "✅ Interaktive Plotly-Dashboards"
    ]
    
    for chart in charts:
        print(chart)
    
    print("\n🚀 BEREIT FÜR:")
    print("  • Präsentations-Slides")
    print("  • Wissenschaftliche Paper")
    print("  • Conference-Posters")
    print("  • Executive-Dashboards")

# Führe alle Visualisierungen aus
if __name__ == "__main__":
    create_all_visualizations()
```
